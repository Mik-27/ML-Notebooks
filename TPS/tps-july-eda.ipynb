```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("/kaggle/input/tabular-playground-series-jul-2021/train.csv")
test = pd.read_csv("/kaggle/input/tabular-playground-series-jul-2021/test.csv")
display(train.head())
display(test.head())
```

```python
cols = list(train.columns)
cols.pop(0)
cols
```

```python
fig = plt.figure(figsize=(10, 6))

sns.histplot(data=train, x='deg_C', kde=True, stat='frequency')
```

```python
fig = px.histogram(train, x='deg_C', marginal='box', nbins=25, opacity=0.7)

# fig.update_traces(nbinsx=0, xbins_size=10)

fig.show()
```

```python
fig = ff.create_distplot([train['deg_C'], train['relative_humidity']], group_labels=['Temperature', 'Relative Humidity'], show_rug=False)
fig.update_layout(template='plotly_white',
                  xaxis=dict(
                        mirror=True,
                        ticks='outside',
                        showline=True,
                  ),
                  yaxis=dict(
                        mirror=True,
                        ticks='outside',
                        showline=True,
                        
                  ),
                 )
fig.show()
```

# Distribution of Training DataðŸ“Š

```python
for col in cols:
    if col.startswith('target'):
        color = '#E45756'
    else:
        color = '#4c78a8'
    fig = ff.create_distplot([train[col]], group_labels=[col], show_rug=False, colors=[color])
    fig.update_layout(showlegend=False, xaxis_title=col, template='plotly_white')
    fig.show()
```

# Distribution of Testing DataðŸ“Š

```python
for col in cols:
    if col.startswith('target'):
        cols.remove(col)
# print(cols)
for col in cols:    
    fig = ff.create_distplot([test[col]], group_labels=[col], show_rug=False)
    fig.update_layout(showlegend=False, xaxis_title=col, template='plotly_white')
    fig.show()
```
